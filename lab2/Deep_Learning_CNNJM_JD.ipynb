# Forward Propagation using Dot Product

```python
input_data = np.array([2,3,2])
weights = {'node_0':np.array([1,1,1]),
           'node_1':np.array([-1,1,1]),
           'node_2':np.array([1,1,1]),
           'output':np.array([2,-1,1])}

node_0_value = (input_data*weights['node_0']).sum()
node_1_value = (input_data*weights['node_1']).sum()
node_2_value = (input_data * weights['node_2']).sum()
```

```python
hidden_layer_values = np.array([node_0_value, node_1_value, node_2_value])
print(hidden_layer_values)
```

```python
output = (hidden_layer_values * weights['output']).sum()
print(output)
```
